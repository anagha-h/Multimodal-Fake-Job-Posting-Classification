```python
import pandas as pd

# Load the dataset
df = pd.read_csv('fake_job_postings.csv')

# Display the first few rows to confirm loading
display(df.head())
```

# Task
Perform undersampling on the data, train a classification model on the undersampled data, and compare the classification results with the results from the original data.

## Prepare data for sampling

### Subtask:
Handle categorical features by encoding them, as sampling techniques require numerical inputs.


**Reasoning**:
Identify categorical columns, apply one-hot encoding, drop original categorical columns, and separate features and target variable.



```python
# Identify categorical columns (excluding the target variable 'fraudulent')
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
if 'fraudulent' in categorical_cols:
    categorical_cols.remove('fraudulent')

# Apply one-hot encoding to categorical columns
df_encoded = pd.get_dummies(df, columns=categorical_cols, dummy_na=False)

# Separate features (X) and target variable (y)
X = df_encoded.drop('fraudulent', axis=1)
y = df_encoded['fraudulent']

# Display the first few rows of X and y to confirm the changes
display(X.head())
display(y.head())
```

## Train and evaluate a classification model on the original data

### Subtask:
Train a basic classification model (e.g., Logistic Regression) on the original, imbalanced data and evaluate its performance.


**Reasoning**:
Train a Logistic Regression model on the original imbalanced data and evaluate its performance using classification report and confusion matrix.



```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Instantiate a LogisticRegression model
# Set max_iter to a higher value for convergence
model = LogisticRegression(max_iter=1000)

# Fit the model to the original data
model.fit(X, y)

# Predict the target variable
y_pred = model.predict(X)

# Generate and print classification report and confusion matrix
print("Classification Report on Original Data:")
print(classification_report(y, y_pred))

print("\nConfusion Matrix on Original Data:")
print(confusion_matrix(y, y_pred))
```

Apply RandomUnderSampler to the preprocessed data to balance the classes by reducing the majority class.


**Reasoning**:
Apply RandomUnderSampler to the preprocessed data to balance the classes by reducing the majority class.



```python
from imblearn.under_sampling import RandomUnderSampler

# Instantiate RandomUnderSampler
rus = RandomUnderSampler(random_state=42)

# Apply the sampler to the features X and target variable y
X_undersampled, y_undersampled = rus.fit_resample(X, y)

# Display the shape of the undersampled data to confirm
print("Shape of original X:", X.shape)
print("Shape of undersampled X:", X_undersampled.shape)
print("Shape of original y:", y.shape)
print("Shape of undersampled y:", y_undersampled.shape)
```
